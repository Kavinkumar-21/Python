```python
# 3.Body Mass Index
''' Output will be like this
Enter the BMI Index:34
Very Overweight '''


#Logical-Code

BMI=int(input("Enter the BMI Index: "))
if (BMI<16):
    print("Severe Thinner")
elif (BMI<=17):
    print("Moderate Thinner")
elif (BMI<=18):
    print("Mild Thinner")
elif (BMI<=25):
    print("Normal")
elif (BMI<=30):
    print("Over weight")
elif (BMI<=35):
    print("Obese Class-1")
elif (BMI<=40):
    print("Obese Class-2")
else:
    print("Obese Class-3")
```

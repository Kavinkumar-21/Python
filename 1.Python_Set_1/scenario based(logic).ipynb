```python
"""1.Scenario: A system checks if a user is eligible to vote based on their age.
 Write logic to ask the user for their age and determine if they are eligible to vote based on whether they are 18 or older."""

#Logical Code


Voters_Age = int(input("Enter your age: "))
if Voters_Age<18:
    print("Your Age is below 18, So your are Not Eligible")
elif Voters_Age>=60:
    print("Senior citizen, Your Age is Eligible")
else:
    print("Your Age is Eligible for Voting")

```

```python
'''3.	Scenario: A company provides employees with a 10% bonus if their salary exceeds $50,000.
 Write logic to determine the bonus amount based on the given salary.'''

#Logical Code

Salary = int(input("Enter the salary: "))
if Salary>=50000:
    bonus = Salary * 0.10
    print("your bonus:", bonus)
else:
    bonus = 0
    print("Not eligible for bonus")
```

```python
'''5.	Scenario: A text-processing tool reverses a given word or sentence for formatting purposes.
 Write logic to take a word or sentence as input and produce its reversed version.'''

#Logical Code

text = input("Enter a word or sentence: ")
reversed_text = text[: :-1]  #slicing method used starting index value 0 and stop value 0 and stepup value -1 so it read it from last index to 0.
print("Reversed text:", reversed_text)
```

```python
'''7.	Scenario: A retail store offers a 20% discount if a customerâ€™s total order exceeds $100. 
Write logic to calculate the final amount to be paid after applying the discount.'''


#Logical Code


order_amount = int(input("Enter the total order amount: "))
if order_amount > 100:
    discount = order_amount * 0.20
    final_amount = order_amount - discount
    print("Discount applied: $", discount)
else:
    final_amount = order_amount
    print("No discount applied.")

print("Final amount to be paid: $", final_amount)
```

```python
'''8.Scenario: A banking system processes withdrawal requests and ensures the user has enough balance.
 Write logic to check if a user has enough balance before allowing a withdrawal and update the remaining balance accordingly.'''

#Logical Code


account_balance = int(input("Enter your current balance: "))
Withdrawal = int(input("Enter the withdrawal amount: "))
if account_balance > Withdrawal:
    Enough_Bal = account_balance - Withdrawal
    print("Withdrawal Success and Your current account balance: ",Enough_Bal)
else:
    print("You don't have a sufficient balance in your account")
```

```python
'''9.Scenario: A calendar system verifies whether a given year is a leap year based on standard leap year rules.
 Write logic to determine whether a given year is a leap year.'''

#Logical Code

year = int(input("Enter a year: "))
if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(year, "is a leap year.")
else:
    print(year, "is not a leap year.")

```
